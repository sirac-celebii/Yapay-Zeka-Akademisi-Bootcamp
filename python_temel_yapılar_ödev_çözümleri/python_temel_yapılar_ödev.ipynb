```python
# ============================================================
# SORU 1
# Bir değişken tanımlayalım: ad = "Kaan", yas = 25, ortalama = 3.45
# Bu değişkenlerin tiplerini type() ile yazdıralım.
# ============================================================

ad = "Kaan"
yas = 25
ortalama = 3.45

print(f"Type of the ad variable         : {type(ad)}")
print(f"Type of the yas variable        : {type(yas)}")
print(f"Type of the ortalama variable   : {type(ortalama)}")
```

```python
# ============================================================
# SORU 2
# Kullanıcıdan yaş bilgisini input() ile alalım.
# Bu yaşın tipini ekrana basalım ve 5 yıl ekleyip sonucu yazdıralım.
# Not: input() her zaman string döndürür, int'e çevirmeyi unutmayalım.
# ============================================================

while True:
    try:
        age = int(input("Yaşınızı giriniz -> "))
        break
    except ValueError as e:
        print("Invalid input, try again !")
        continue

result = age + 5

print(f"Age                      : {age}")
print(f"Type of the age variable : {type(age)}")
print("="* 60)

print(f"Result                   : {result}")
```

```python
# ============================================================
# SORU 3
# Bir ürün fiyatı (float) alalım. %18 KDV hesaplayalım.
# Toplam fiyatı 2 basamak olacak şekilde yazdıralım.
# ============================================================

product_price = 100

tax = 100 * 0.18

total_price = tax + product_price

print(f"Product price : {product_price}")
print(f"Tax           : {tax}")
print("="* 60)

print(f"Total price   : {total_price:.2f}")
```

```python
# ============================================================
# SORU 4
# Bir liste oluşturalım: sayilar = [10, 20, 30, 40, 50]
# - İlk elemanı yazdıralım
# - Son elemanı yazdıralım
# - 2. indexten sona kadar olan parçayı yazdıralım
# - Listeye 60 ekleyelim
# - Listedeki 20 değerini silelim
# ============================================================

sayilar = [10, 20, 30, 40, 50]

print(f"First element               : {sayilar[0]}")
print(f"Last element                : {sayilar[-1]}")
print(f"Second index to last index  : {sayilar[2:]}")
print("="* 60)

sayilar.append(60)
print(f"60 added                   : {sayilar}")
print("="* 60)

sayilar.remove(20)
print(f"20 removed                 : {sayilar}")
```

```python
# ============================================================
# SORU 5
# Bir tuple oluşturalım: koordinat = (12, 34)
# - Tuple içindeki değerleri unpacking ile x ve y değişkenlerine alalım
# - x ve y'yi yazdıralım
# - Tuple'ın değiştirilemediğini göstermek için (yorum satırıyla) örnek verelim
# ============================================================

coordinates = (12,24)

coor_x, coor_y = coordinates

print(f"Coordinate of x : {coor_x}")
print(f"Coordinate of y : {coor_y}")
print("="* 60)

# Lists are mutable(we can change a list's elements)
my_list = [12, 24]
my_list.append(36)
print(f"36 added to my_list : {my_list}")
print("="* 60)

# Tuples are indexable like a lists
print(f"First element of the tuple : {coordinates[0]}")
print("="* 60)

# Tuples are immutable(we can't change a tuple's elements) unlike lists
# coordinates.append(36) -> Raises an error
# coordinate[0]           -> Raises an error
```

```python
# SORU 6
# Bir sözlük (dictionary) oluşturalım:
# ogrenci = {"isim": "Ayşe", "yas": 22, "bolum": "Yazılım"}
# - Öğrencinin ismini yazdıralım
# - "not" anahtarı ile 90 ekleyelim
# - "yas" değerini 23 yaparak güncelleyelim
# - Tüm anahtarları ve tüm değerleri yazdıralım
# ============================================================

ogrenci = {"isim": "Ayşe", 
           "yas": 22, 
           "bolum": "Yazılım"
           }

print(f"Name of the student : {ogrenci["isim"]}")
print("="* 60)

ogrenci.update({"not" : 90})
print(f"'not' added : {ogrenci}")
print("="* 60)

ogrenci["yas"] = 23
print(f"'yas' has changed to 23 : {ogrenci}")
print("="* 60)
for key, value in ogrenci.items():
    print(f" Key : {key} | Value : {value}")
```

```python
# ============================================================
# SORU 7
# Set oluşturalım ve tekrar edenleri temizleyelim:
# liste = ["Ali", "Ayşe", "Ali", "Mehmet", "Ayşe"]
# - listeyi set'e çevirip benzersiz isimleri yazdıralım
# - benzersiz isim sayısını yazdıralım
# ============================================================

liste = ["Ali", "Ayşe", "Ali", "Mehmet", "Ayşe"]
liste = set(liste)
print(f"Unique Names : {liste}")
print("=" * 60)

print(f"Unique name count : {len(liste)}")
```
