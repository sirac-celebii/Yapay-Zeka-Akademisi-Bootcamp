```python
# ============================================================
# SORU 1 (IF)
# Kullanıcıdan bir sayı alın.
# Sayı pozitifse "Pozitif", negatifse "Negatif", sıfırsa "Sıfır" yazdırın.
# ============================================================
while True:
    try:
        sayi = int(input("Bir sayi giriniz -> "))
        break
    except ValueError as e:
        print("Lütfen sadece sayı giriniz !")
        continue

if sayi > 0:
    print("Girdiğiniz sayı Pozitif bir sayı")
elif sayi < 0:
    print("Girdiğiniz sayı Negatif bir sayı")
else:
    print("Girdiğiniz sayı Sıfır")
```

```python
# ============================================================
# SORU 2 (FOR)
# 1'den 10'a kadar (10 dahil) sayıları yazdırın.
# Ayrıca bu sayıların toplamını hesaplayıp ekrana yazdırın.
# ============================================================

toplam = 0

for i in range(1, 11):
    print(i)
    toplam += i

print("=" * 60)
print(f"1'den 10'a kadar olan sayıların toplamı : {toplam}")
```

```python
# ============================================================
# SORU 3 (WHILE)
# Kullanıcıdan "q" yazana kadar sürekli giriş alın.
# Kullanıcı her giriş yaptığında "Girdiniz: ..." şeklinde ekrana yazdırın.
# Kullanıcı "q" yazarsa döngü bitsin ve "Çıkış yapıldı" yazsın.
# ============================================================

girdi = None

while girdi != "q":
    girdi = input("Bir harf giriniz -> ")
    print(f"Girdiğiniz harf : {girdi}")

print("Çıkış yapıldı !")
```

```python
# ============================================================
# SORU 4 (NESTED)
# 1'den 20'ye kadar sayıları dolaşın.
# Eğer sayı çiftse "Çift", tekse "Tek" yazdırın.
# Ayrıca sayı 10'dan büyükse yanına "Büyük", değilse "Küçük/Eşit" yazdırın.
# Örnek çıktı: 12 -> Çift - Büyük
# ============================================================

for i in range(1, 21):
    if i % 2 == 0:
        if i > 10:
            print(f"{i} sayısı Çift ve 10'dan Büyüktür")
        elif i <= 10:
            print(f"{i} sayısı Çift ve 10'dan Küçüktür/Eşittir")
    else:
        if i > 10:
            print(f"{i} sayısı Tek ve 10'dan Büyüktür")
        elif i <= 10:
            print(f"{i} sayısı Tek ve 10'dan Küçüktür")                
```
